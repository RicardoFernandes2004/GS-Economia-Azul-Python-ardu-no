```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# dados dos níveis de petróleo
df_niveis = pd.read_csv('nivel_petroleo.csv')
df_niveis['timestamp'] = pd.to_datetime(df_niveis['timestamp'], format='%Y-%m-%d %H:%M:%S')
```

```python
df_niveis
```

```python
# Plotar níveis de petróleo
plt.figure(figsize=(10, 5))
plt.plot(df_niveis['timestamp'], df_niveis['nivel_petroleo'], label='Nível de Petróleo')
plt.xlabel('Timestamp')
plt.ylabel('Nível de Petróleo (%)')
plt.title('Nível de Petróleo ao Longo do Tempo')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
